# Ejercicio 1:
1. Crear una lista con los números mayores que 10.

2. Crear otra lista con los números pares.

3. Calcular la suma de los números impares.

4. Detectar si existe algún número mayor que 50.

```python
numbers = [12, 5, 18, 7, 30, 2, 25, 78, 56]
corte = [x for x in numbers if x > 10]
print(corte)
```

```python
pares = [x for x in numbers if x % 2 == 0]
print(pares)
```

```python
import numpy as np
impares = [x for x in numbers if x % 2 != 0]
suma = np.sum(impares)
print(suma)
```

```python
over50 = [x for x in numbers if x > 50]

if any(num > 50 for num in numbers):
  print(over50)
else:
  print("No number grater than 50")
```

#Ejercicio 2
Crear una lista donde:

Si la temperatura > 25 → añadir "HOT"

Si < 15 → "COLD"

Si no → "OK"

Contar cuántos días hay de cada tipo.

```python
temps = [15, 22, 30, 18, 27, 10]
hot_temps_list = [x for x in temps if x > 25]
cold_temps_list = [x for x in temps if x < 15]
ok_temps_list = [x for x in temps if x >= 15 and x <= 25]


print(f"Hot days: {hot_temps_list}")
print(f"Cold days: {cold_temps_list}")
print(f"OK days: {ok_temps_list}")
```

#Ejercicio 3
- Filtrar usuarios activos mayores de edad.
- Crear una lista con los nombres de usuarios que han gastado más de 100.
- Calcular el total de compras de usuarios activos.
- Crear un diccionario donde:
  - clave = nombre
  - valor = categoría:
    - "VIP" si purchases > 200
    - "Regular" si purchases entre 100 y 200
    - "Low" si < 100

```python
users = [
    {"name": "Alice", "age": 25, "active": True, "purchases": 300},
    {"name": "Bob", "age": 17, "active": False, "purchases": 50},
    {"name": "Charlie", "age": 30, "active": True, "purchases": 120},
    {"name": "Diana", "age": 22, "active": True, "purchases": 80},
]

over_age = [
    user for user in users if user["active"] and user["age"] > 18
]
print(over_age)

over_100 = [
    user for user in users if user["active"] and user["purchases"] > 100
]
print(over_100)

active = [
    user for user in users if user["active"]
]

print(active)
print(len(active))

total = sum(user["purchases"] for user in active)
print(total)


categories = {
    user["name"]: (
        "VIP" if user["purchases"] > 200
        else ("Regular" if 100 <= user["purchases"] <= 200 else "Low")
    )
    for user in users
}
print(categories)
```

#Ejercicio 5
- Crear un diccionario con el total de ventas por ciudad.
- Detectar la ciudad con más ventas.
- Crear una lista ordenada de ciudades por total vendido (descendente).
- Calcular el promedio de ventas por ciudad.

```python
sales = [
    {"city": "Madrid", "amount": 200},
    {"city": "Barcelona", "amount": 150},
    {"city": "Madrid", "amount": 300},
    {"city": "Valencia", "amount": 100},
    {"city": "Barcelona", "amount": 50},
]

sales_by_city = {}
for sale in sales:
  city = sale["city"]
  amount = sale["amount"]
  sales_by_city[city] = sales_by_city.get(city, 0) + amount

print(f"Total sales by city: {sales_by_city}")

max_city = max(sales_by_city, key=sales_by_city.get)

print(f"City with the most sales: {max_city}")
```
