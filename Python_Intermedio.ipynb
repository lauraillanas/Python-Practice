### convertir diccionarios en pandas

```python
experimentos = {
    "exp1": {"temp": 37, "rendimiento": 0.82},
    "exp2": {"temp": 30, "rendimiento": 0.65},
    "exp3": {"temp": 42, "rendimiento": 0.90}
}

#Convierte el diccionario 'experimentos' en un dataframe:
df = pd.DataFrame(experimentos).T # Added .T to transpose the DataFrame
print(df)

#filtra los experimentos con rendimiento > 0.8
filtrado = {k for k, v in experimentos.items() if v["rendimiento"] > 0.8}
print(filtrado)

#Añade una columna 'exito' (True si rendimiento > 0.75):
df["exito"] = df["rendimiento"] > 0.75
print(df)

#Calcula media, máximo y mínimo de rendimiento
media = df["rendimiento"].mean()
maximo = df["rendimiento"].max()
minimo = df["rendimiento"].min()
print(f"Media: {media}, Máximo: {maximo}, Mínimo: {minimo}")

#Ordena por temperatura descendente:
df_ordenado = df.sort_values(by="temp", ascending=False)
print(df_ordenado)

#Exporta el dataframe a CSV
df.to_csv('experimentos.csv', index=False)
print("DataFrame 'df' exportado a 'experimentos.csv'")

```

```python
#Simula datos de expresión génica para 100 genes
genes = [f"GENE_{i}" for i in range(1, 101)]

import random
import builtins # Import builtins to access original functions

expresion = {
    gen: round(random.lognormvariate(1, 0.5), 2)
    for gen in genes
}

print(expresion)

min_val = builtins.min(expresion.values())
max_val = builtins.max(expresion.values())
print(min_val, max_val)

#Para añadir más realismo, añadir genes housekeeping, genes muy expresados y genes silenciados:
for gen in genes[:10]:          # housekeeping
    expresion[gen] *= 2

for gen in genes[10:20]:        # sobreexpresados
    expresion[gen] *= 5

for gen in genes[20:30]:        # casi silenciados
    expresion[gen] *= 0.2

#añade tres columnas al dataframe con estos genes y escribir true or falso dado el caso:

```

```python
import pandas as pd

# Crear un DataFrame a partir del diccionario de expresión
df_genes = pd.DataFrame.from_dict(expresion, orient='index', columns=['expresion_value'])

# Inicializar las nuevas columnas booleanas a False
df_genes['housekeeping'] = False
df_genes['sobreexpresado'] = False
df_genes['silenciado'] = False

# Identificar los genes por categoría y actualizar las columnas
# Los primeros 10 genes son housekeeping
df_genes.loc[genes[:10], 'housekeeping'] = True

# Los genes del 10 al 19 son sobreexpresados (índice 10 a 19)
df_genes.loc[genes[10:20], 'sobreexpresado'] = True

# Los genes del 20 al 29 son silenciados (índice 20 a 29)
df_genes.loc[genes[20:30], 'silenciado'] = True

# Mostrar las primeras filas del DataFrame con las nuevas columnas
display(df_genes.head(35))
```

## 4. Lógica, flujo de control y filtrado

```python
import pandas as pd
orders = [
    {"status": "paid", "total": 120, "customer": {"premium": True}},
    {"status": "cancelled", "total": 200, "customer": {"premium": True}},
    {"status": "paid", "total": 80, "customer": {"premium": False}},
]

print(orders)
```

###
    1.	Filtrar pedidos válidos
	2.	Ordenarlos por cliente premium y total
	3.	Calcular el total final

```python
import pandas as pd

paid_orders = []

for order in orders:
    if order["status"] == "paid":
        paid_orders.append(order)

print(paid_orders)


sorted_orders = sorted(paid_orders, key=lambda x: (not x['customer']['premium'], x['total']))
print(sorted_orders)

total_sum = 0
for order in sorted_orders:
    total_sum += order['total']

print(f"Total sum of paid orders: {total_sum}")

#imprime el resultado en formato CSV
df = pd.DataFrame(sorted_orders)
df.to_csv('orders.csv', index=False)
print("DataFrame 'df' exportado a 'orders.csv'")
```
