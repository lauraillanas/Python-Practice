### Revisar si hay columnas mal tipadas

```python
mal_typed_cols = {}

for col in categorical_cols:
    try:
        # Try to convert to numeric
        if pd.to_numeric(df[col], errors='coerce').notna().sum() / len(df) > 0.8:
            mal_typed_cols[col] = 'numeric'
        # Try to convert to datetime
        elif pd.to_datetime(df[col], errors='coerce').notna().sum() / len(df) > 0.8:
            mal_typed_cols[col] = 'datetime'
    except:
        pass

if mal_typed_cols:
    print("Columnas que podrían estar mal tipadas (y su tipo sugerido):")
    for col, dtype in mal_typed_cols.items():
        print(f"- '{col}': sugiero '{dtype}'")
else:
    print("No se encontraron columnas que parezcan estar mal tipadas significativamente en las columnas categóricas.")


# Additionally, let's check for 'object' columns that are actually boolean
boolean_cols_as_object = []
for col in categorical_cols:
    unique_values = df[col].unique()
    if len(unique_values) <= 2 and all(v in ["Yes", "No", True, False, "True", "False", 0, 1] for v in unique_values):
        boolean_cols_as_object.append(col)

if boolean_cols_as_object:
    print("\nColumnas 'object' que podrían ser booleanas:")
    for col in boolean_cols_as_object:
        print(f"- {col}")
else:
    print("\nNo se encontraron columnas 'object' que parezcan ser booleanas.")
```

Calcula métricas como:

```python
#Calcula el promedio del tiempo de uso
average_time_usage = df['average_session_length_minutes'].mean()
print(f"Promedio de tiempo de uso: {average_time_usage:.2f} horas")
```

```python
# Definir los rangos de edad
bins = [0, 18, 25, 35, 45, 55, 65, df['age'].max() + 1] # Sumamos 1 para asegurar que el último bin sea único
labels = ['<18', '18-24', '25-34', '35-44', '45-54', '55-64', '65+']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Calcular el promedio de tiempo de sesión por grupo de edad
session_length_by_age_group = df.groupby('age_group')['average_session_length_minutes'].mean().reset_index()

plt.figure(figsize=(12, 7))
sns.barplot(x='age_group', y='average_session_length_minutes', data=session_length_by_age_group, palette='viridis')
plt.title('Tiempo Promedio de Sesión por Grupo de Edad')
plt.xlabel('Grupo de Edad')
plt.ylabel('Tiempo Promedio de Sesión (minutos)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Ejercicio 5

###Heatmap de correlación entre variables numéricas y gráfico que compare engagement por país

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Heatmap de correlación entre variables numéricas
numeric_cols = df.select_dtypes(include=['number']).columns

# Eliminar columnas con todos los valores NaN (si hubiera alguna después de la limpieza)
numeric_df = df[numeric_cols].dropna(axis=1, how='all')

# Calcular la matriz de correlación
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap de Correlación entre Variables Numéricas')

# Limpiar las etiquetas de los ejes del heatmap
cleaned_numeric_cols = [col.replace('_', ' ').title() for col in numeric_cols]
plt.xticks(ticks=range(len(numeric_cols)), labels=cleaned_numeric_cols, rotation=90, ha='right')
plt.yticks(ticks=range(len(numeric_cols)), labels=cleaned_numeric_cols)

plt.show()

# 2. Gráfico que compare engagement por país (Top 10 países)
engagement_by_country = df.groupby('country')['user_engagement_score'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=engagement_by_country.index, y=engagement_by_country.values, palette='plasma')
plt.title('Top 10 Países con Mayor Puntuación de Engagement de Usuario')
plt.xlabel('País')
plt.ylabel('Puntuación de Engagement Promedio')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
